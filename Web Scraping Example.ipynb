### First, import some packages
The below is the most useful arrangement I've found

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Let's see if we can scrape the urls from 'Top 5 Data Science and Machine Learning Course for Programmers' by DZone
url='https://dzone.com/articles/top-5-data-science-and-machine-learning-course-for'
```

```python
site=requests.get(url)
```

And if we take a look at what that returns

```python
print(type(site))
```

```python
print(site.text)
```

```python
soup=bs(site.text,"html")
```

```python
all_a=soup.find_all('a')
print(all_a)
```

```python
urls=[link.get('href') for link in all_a]
```

```python
urls
```

For exporting or playing around, it's usually best to put these in pandas

```python
#If you input a dictionary, it will parse the key as the column header
url_df=pd.DataFrame({'url':urls})
```

```python
url_df.head()
```

```python
#Remove na's
url_df_full=url_df[url_df['url'].isnull()==False]
```

```python
url_df_full.head()
```

```python
links_df=url_df_full[url_df_full['url'].str.contains('http')]
print(links_df.head())
```

```python
links_df.to_csv('Scraped Links',index=False)
```
